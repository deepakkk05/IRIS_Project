```python
import pandas as pd
df = pd.read_csv(r"C:\Users\91971\Downloads\IRIS.csv")


df['species'] = df['species'].astype('category').cat.codes

corr = df.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')

# Show the plot
plt.show()
```
